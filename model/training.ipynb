```python
import cv2, os
data_path = os.getcwd() + '\\dataset\\train'
data_path = data_path.replace("\\model", "")
categories= os.listdir(data_path)
labels = [i for i in range(len(categories))]

label_dict = dict(zip(categories,labels)) #empty dictionary
```

```python
img_size=150
data=[]
target=[]

for c in categories:
    if c == ".DS_Store":
        continue
    else:
        folder_path=os.getcwd() + '\\dataset\\train\\'+ c
        folder_path =  folder_path.replace("\\model", "")
        img_names=os.listdir(folder_path)

        for iName in img_names:
            img_path=os.path.join(folder_path, iName)
            i=cv2.imread(img_path)

            try:
                resized=cv2.resize(i, (img_size, img_size))
                data.append(resized)
                target.append(label_dict[c])
            except Exception as e:
                print("Did not work loser, here: ")


    
```

```python
import numpy as np
data=np.array(data)/255.0
data=np.reshape(data,(data.shape[0], img_size, img_size, 3))
target=np.array(target)

from keras.utils import to_categorical

newTarget = keras.utils.to_categorical(target)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ModelCheckpoint

# Initialize model
model=Sequential()

#First CNN (convolutional neural network) Layer
model.add(Conv2D(234, (3,3), activation='relu', input_shape=data.shape[1:]))
model.add(MaxPooling2D(pool_size=(2,2)))

# Second convolution layer to get better accuracy on the images
model.add(Conv2D(100, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

#Flatten layer to stack the out convolution from second layer
model.add(Flatten())

# Dense layer of 128 neurons
model.add(Dense(128,activation='relu'))

#Final Layer with 3 outputs for the 3 categories (garbage, organic, recycling)
model.add(Dense(3, activation='softmax'))


model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_target, test_target=train_test_split(data,newTarget,test_size=0.1)
```
