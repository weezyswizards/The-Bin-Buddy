```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_path = os.getcwd() + '\\dataset\\train'
train_path = train_path.replace("\\model", "")
valid_path = os.getcwd() + '\\dataset\\validation'
valid_path = valid_path.replace("\\model", "")

# Assuming data is in 'train' directory with subdirectories for each class
datagen = ImageDataGenerator(validation_split=0.2, rescale=1./255)

# Load and preprocess training data
train_generator = datagen.flow_from_directory(train_path,
    target_size=(150, 150),  # Resize images to 100x100 pixels
    batch_size=32,
    class_mode='sparse',  # Use 'sparse' mode for sparse_categorical_crossentropy
    subset='training'
)

# Load and preprocess validation data
validation_generator = datagen.flow_from_directory(valid_path,
    target_size=(150, 150),  # Resize images to 100x100 pixels
    batch_size=32,
    class_mode='sparse',  # Use 'sparse' mode for sparse_categorical_crossentropy
    subset='validation'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ModelCheckpoint

# Initialize model
model=Sequential()

#First CNN (convolutional neural network) Layer
model.add(Conv2D(234, (3,3), activation='relu', input_shape=data.shape[1:]))
model.add(MaxPooling2D(pool_size=(2,2)))

# Second convolution layer to get better accuracy on the images
model.add(Conv2D(100, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

#Flatten layer to stack the out convolution from second layer
model.add(Flatten())

# Dense layer of 128 neurons
model.add(Dense(50,activation='relu'))

#Final Layer with 3 outputs for the 3 categories (garbage, organic, recycling)
model.add(Dense(3, activation='softmax'))


model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```
